```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

# The subplot() function is used to create multiple plots within a single figure. 
# It takes three arguments: the number of rows, the number of columns, and the index of the plot. 
# The index starts at 1 and increases from left to right and top to bottom.

# For example, if we use subplot(2, 2, 1), it will create a figure with two rows and two columns, and the first plot will be in the top left corner

import numpy as np
import matplotlib.pyplot as plt

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the figure and the subplots
fig, axs = plt.subplots(2, 2)

# Plot the data on each subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Add a main title to the figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

# A bar plot, also known as a bar chart or a column chart, is a type of chart that represents data with rectangular bars.
# The length or height of each bar is proportional to the value it represents.

# Bar plots are useful for comparing different categories or groups of data. 
# They can be used to show the distribution of a single categorical variable or to compare the values of multiple variables.

# In a vertical bar plot, the x-axis represents the categories or groups, while the y-axis represents the values. 
# In a horizontal bar plot, the x-axis represents the values, and the y-axis represents the categories or groups.

# Bar plots can be created using various Python libraries such as Matplotlib, Seaborn, Plotly, and Pandas. 
# They can be customized with colors, labels, titles, and other properties to improve their visual appearance and readability.


import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("bar plot")
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

# A box plot is a graphical representation of a dataset that shows the distribution of the data through its quartiles. 
# It is also known as a box and whisker plot because it contains a box with whiskers that extend from the box to indicate variability outside the upper and lower quartiles.

# Box plots are used to visualize the spread of data, identify outliers, and compare distributions between different groups of data.
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

plt.show()
```
